## 2. Проверим датафрейм на наличие пропусков:

Необходимо изучить датафрейм на наличие или отсутствие пропусков. Если они есть и составляют большую долю от числа данных - нужно принять решение: заменить ли их каким-либо значением, или удалить. Если доля совсем маленькая, то пропуски можно либо игнорировать, либо попробовать чем-либо заменить.  

Представим в виде объединённой таблицы количество и долю пропусков для наглядности.
Используем функцию лямбда, чтобы не перегружать память.
Сoncat соединяет таблицу с суммой и таблицу со средними значениями фрейма данных.

```python
pd.concat([df.isna().sum(), df.isna().mean().apply(lambda x: '{0:.1f}%'.format(x * 100))], 
          axis=1).rename(columns={0:'количество пропусков', 1:'доля пропусков'})
```

* Как видим, где-то доля пропусков ничтожно мала, переменные категориальные.
* В столбцах, связанных с бюджетом и деньгами - доля пропусков огромная, восстановить данные мы не можем. 
* Есть пропуски в столбцах `film_studio, production_country, director, producer, genres`, их мы заменим на "неизвестно" - сохраняя русско-язычный стиль написания. 
* Есть пропуски в стобце рейтинг, их мы заменим на '0'.
* Удалять мы ничего не будем. 

```python
#составим список колонок, где нужно заменить "упущенные" данные
columns = ['film_studio', 'production_country', 'director', 'producer', 'genres']
#пишем цикл, он перебирает строки в колонках датафрейма и заменяет NAN Значения на "неизвестно"
for omission in columns:
    df[columns] = df[columns].fillna('неизвестно')
#заменим пропуски в столбце рейтинга на пустую строку:
df['ratings'] = df['ratings'].fillna(0)

pd.concat([df.isna().sum(), df.isna().mean().apply(lambda x: '{0:.1f}%'.format(x * 100))], 
          axis=1).rename(columns={0:'кол-во пропусков', 1:'доля пропусков'})
```

Как видим, пропуски в данных остались только в стобцах, связанных с деньгами.

```python
print(df[df['refundable_support'].notna()]['box_office'].isna().sum())
print(df[df['nonrefundable_support'].notna()]['box_office'].isna().sum())
```

```python
df_supported['roi'] = df_supported['box_office'] / df_supported['full_sup'] - 1 #рентабельность

df_supported['is_paid_off'] = df_supported['roi'] > 0 #окупились

df_supported['is_not_paid_off'] = ~df_supported['is_paid_off'] #неокупились
#lambda x: 0 - x.sum() - эта часть опускает не окупившиеся фильмы ниже нуля, подсмотрела в стаковерфлоу
df_supported.groupby('genre') \
            .agg({'title': 'count',
                  'is_paid_off': 'sum',
                  'is_not_paid_off': lambda x: 0 - x.sum()}) \
            .sort_values(by='title',
                         ascending=False) \
            .plot(title='Соотношение окупившихся к не окупившимся фильмам по жанрам',
                  kind='bar',
                  stacked=True,
                  y=['is_paid_off', 'is_not_paid_off'],
                  figsize=(15,4),
                  rot=15);
```
